```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
```

```python
# Sample function call: func2(2, 4)
import time
start = time.time()
print(func2(2, 4))
end = time.time()
result = (end-start)*1000
print("Run Time : " , result)
# The Function performs Multiplication Using Iteration
```

```python
# Exercise 3
def func3(num):
   result = 0
   while num > 0:
    result = result + num % 10
    num //= 10
   return result
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
import time

start = time.time()
print(func4([1, 2, 3, 4], [2, 4, 6, 8]))
end = time.time()
result = (end-start)*1000
print(f"Run Time : {result}")
# This Function is used to find the Number of Common Values Present in Both of the list
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n: # 16
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)
    return A
def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - i + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
    elif n > 2:
        m = n // 3
        func8_helper(A, i, j - m)
        func8_helper(A, i + m, j)
        func8_helper(A, i, j - m)

```

```python
#Sample function call: func8([3,1,2,7,8,6])
import time

start = time.time()
print(func8([3,1,2,7,8,6]))
end = time.time()
result = (end-start)*1000
print(f"Run Time : {result}")
# The Function Performs the Sorting array in ascending order
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
import time

start = time.time()
print(func12([1, 2, 3, 4, 5], 2))
end = time.time()
result = (end-start)*1000
print(f"Run Time : {result}")
# This Function is used to take last k elements from the list , and then push then in the front of the list.
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
  even = 0
  odd = 0
  for i in arr:
    if(i%2==0):
      even = even + 1
    else:
      odd = odd+1
  print(f"Even Number Count : {even}")
  print(f"Odd Number Count : {odd}")


```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr,key):
  for i in range(len(arr)):
    if arr[i] == key:
      return print(f"Element Found at Index : {i}")
  return print("Element Not Found ")
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
import time
arr = [1, 3, 5, 7, 9]
start = time.time()
linear_Search(arr,5)
end = time.time()
result = (end-start)*1000
print(f"Run Time : {result}")
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr,key):
  low = 0 ;
  high = len(arr) - 1
  while low<=high:
    mid = (low+high)//2
    if arr[mid] == key:
      return print(f"Element Found At Index : {mid} ")
    elif arr[mid] < key:
      low = mid + 1
    else:
      high = mid - 1
  return -1

```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
import time
arr = [1, 3, 5, 7, 9,15,16,14,45]
start = time.time()
Binary_Search(arr,7)
end = time.time()
result = (end-start)*1000
print(f"Run Time : {result}")
```
